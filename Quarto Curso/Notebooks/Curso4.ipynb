# Manipulando os dados

## Criando colunas numéricas

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=';')
dados.head()
```

```python
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio'] # Cria uma nova Coluna com a soma do valor Mensal aluguel + condominio
dados.head()
```

```python
dados['Valor_por_ano'] = dados['Valor_por_mes'] * 12 + dados['IPTU'] # O valor por ano se trata de 12 vezes o valor por mes com o adicional do iptu
dados.head()
```

## Criando colunas categóricas

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] # Formata os dados do tipo string em uma nova coluna formando uma descricao
dados
```

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                                        dados['Quartos'].astype(str) + ' quarto(s) ' + \
                                        ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
dados.head() # Formata os dados int em string e forma uma coluna tipo string formando uma descricao
```

```python
dados['Possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else 'Não') # Pega todos os dados da coluna e forma um IfElse '.apply(lambda x:)'
dados.head()
```

```python
dados.to_csv('Dados_filtrados.csv', index=False, sep=';')
```
